```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os

import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.utils import plot_model

from tensorflow.keras.datasets import mnist
from tensorflow.keras.utils import to_categorical


(train_x, train_y), (test_x, test_y) = mnist.load_data()
train_y = to_categorical(train_y) 
test_y = to_categorical(test_y)
train_x = train_x.reshape(train_x.shape[0], 28, 28, 1).astype('float32')/255
test_x = test_x.reshape(test_x.shape[0], 28, 28, 1).astype('float32')/255
tf.config.list_physical_devices('GPU')
```

## 3.1 
(10%) You are asked to construct a classification model based on convolutional neural networks for digit recognition. Please report the prediction accuracy for the testing set.


```python
train_x.shape
train_y.shape
```

## train_參數

```python
e=30 #epochs
batch_size=500
lr=0.001
```

## model架構

```python
inputs = keras.Input(shape=(28, 28, 1))
x = layers.Conv2D(64, (3, 3), activation='relu')(inputs)
x = layers.MaxPool2D()(x)
x = layers.Conv2D(128, (3, 3), activation='relu')(x)
x = layers.Conv2D(64, (3, 3), activation='relu')(x)
x = layers.Flatten()(x)
x = layers.Dense(64, activation='relu')(x)
outputs = layers.Dense(10, activation='softmax')(x)
model = keras.Model(inputs, outputs)
print(model.summary())
```

```python
import os 
model_dir='./model'
log_dir=os.path.join('Q3_logs',model_dir)
os.makedirs(model_dir, exist_ok=True)
os.makedirs(log_dir, exist_ok=True)
model_cbk=keras.callbacks.TensorBoard(log_dir=log_dir)
model_mckp=keras.callbacks.ModelCheckpoint(model_dir+'/Best_model.h5',save_best_only=True)
```

```python


model.compile(keras.optimizers.Adam(learning_rate=lr), 
           loss=keras.losses.CategoricalCrossentropy(), 
           metrics=[keras.metrics.CategoricalAccuracy()])

# 訓練網路模型
CNN_history=model.fit(train_x,train_y,batch_size=batch_size,
          epochs=e
          ,validation_data=(test_x,test_y)
                      # ,
          # callbacks=[model_cbk, model_mckp]
         )
```

```python
CNN_history.history.keys()
```

```python
# plt.plot(CNN_history.history['val_categorical_accuracy'],label='test')
plt.plot(CNN_history.history['categorical_accuracy'],label='train')
# plt.text(20, 20,)
plt.title('Accuracy :train_acc:{:.4f}  test_acc:{:.4f}'.format(CNN_history.history['categorical_accuracy'][-1],CNN_history.history['val_categorical_accuracy'][-1]))

plt.legend()
plt.savefig('3_1_CNNaccuracy.jpg')
```

```python
# plt.plot(CNN_history.history['val_loss'],label='test')
plt.plot(CNN_history.history['loss'],label='train')
plt.title('Loss :train_acc:{:.4f}  test_acc:{:.4f}'.format(CNN_history.history['loss'][-1],CNN_history.history['val_loss'][-1]))

plt.legend()
plt.savefig('3_1_CNNloss.jpg')
```

```python
res = tf.keras.metrics.CategoricalAccuracy()
res.update_state(test_y, model.predict(test_x))
print("test_acc:",res.result().numpy())
```

```python
model.evaluate(train_x,train_y)
```

```python
import random
import numpy as np

def add_noise(raw_data,noise_lv,img_size):
    random.seed(10)
    noise_data=raw_data.copy()
    for i in range(len(noise_data)):
        ran_seq = random.sample([n for n in range(img_size*img_size)], np.int(img_size*img_size*noise_lv)) #隨機選取比例的pixel
        x = noise_data[i].copy().reshape(-1, img_size*img_size) 
        x[0, ran_seq]=1 # 取代為255/255
        noise_data[i,:]=x.reshape(-1, img_size, img_size, 1)
    return noise_data


test_x_n10 =  add_noise(test_x,0.1,28)
test_x_n20 =  add_noise(test_x,0.2,28)
test_x_n30 =  add_noise(test_x,0.3,28)

# def get_noise_raw():
#     (_,_ ), (test_x_for_n,_) = mnist.load_data()
#     test_x_for_n = test_x.reshape(test_x_for_n.shape[0], 28, 28, 1).astype('float32')/255
#     return test_x_for_n
# # new_img = np.empty((len(img), size, size, ch))
# # new_img[i,:] = x.reshape(-1, size, size, ch)
```
